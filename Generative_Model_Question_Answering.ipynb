```python
import time

start = time.time()

file1 = open('COVID-Dialogue-Dataset-English.txt', 'r')
Lines = file1.readlines()
strings= []
position = 0
for i, line in enumerate(Lines):
    if i< position:
        continue
    if line.startswith('Dialogue'):
        position = i
        sub_count = 1
        string = ''
        while (i+sub_count<len(Lines)) and (not Lines[i+sub_count].startswith('id=')):
            string += Lines[i+sub_count]
            sub_count += 1
            position += 1
        string = string.replace('Thanks for your question on Healthcare Magic.', '')
        string = string.replace('In brief:', '')
        string = string.replace('Would you like to video or text chat with me?', '')
        string = string.replace('Thanks for using Healthcaremagic.com', '')
        string = string.replace('I can understand your concern.', '')
        string = string.replace('I have reviewed your query and here is my advice.', '')
        string = string.replace('Warm  welcome to Healthcaremagic.com', '')
        string = string.replace('Respected user', '')
        string = string.replace('Hello', '')
        string = string.replace('https://www.healthtap.com/blog/covid-19-care-guidelines/preventing-covid-19', '')
        strings.append(string)
strings
```

```python
f = open("questions_and_responses.txt", "w")
for i in range(len(strings)):
    f.write(strings[i])
    f.write('\n')
end = time.time()
print(end-start)
```
